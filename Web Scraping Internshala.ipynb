# Importing Important Libraries

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Retrieving the content of the web page located at the specified URL

```python
webpage = requests.get('https://internshala.com/internships/matching-preferences/')
```

# Using BeautifulSoup, it provides various methods and properties to navigate, search, and extract information from the webpage structure

```python
html_content = webpage.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
companies = soup.find_all('div', class_ = 'internship_meta')
```

# Scraping Data

```python
comp = []
profile = []
place = []
start_date = []
salary = []
exp_req = []
posted_date = []

for company in companies:
    
    # Company Name
    comp.append( company.find('a', class_ = 'link_display_like_text').text.strip() )
    
    # Job Profile
    profile.append( company.find('a', class_ = 'view_detail_button').text.strip() )

    # Place of Posting
    place.append( company.find('a', class_ = 'location_link').text.strip() )

    # Start Date
    start_date.append( company.find('div', class_ = 'item_body').text.split()[1] )

    # Salary
    salary.append( company.find('div', class_ = 'individual_internship_details').find_all('div', class_ = 'item_body')[1].text.strip() )

    # Experience Required
    exp_req.append( company.find('div', class_ = 'individual_internship_details').find_all('div', class_ = 'item_body')[2].text )

    # Posted Date
    posted_date.append( company.find('div', class_ = 'status-container').text.strip() )
    

    
scraped_df = pd.DataFrame({'Company Name':comp, 'Job Profile':profile, 'Start_Date':start_date, 'Salary':salary, 'Experience Required':exp_req, 'Posted Date':posted_date})
```
